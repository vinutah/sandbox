```python
from graphviz import Digraph
```

```python
dot = Digraph(comment='The Round Table')
dot.node('A', 'King Arthur')
dot.node('B', 'Sir Bedevere the Wise')
dot.node('L', 'Sir Lancelot the Brave')

dot.edges(['AB', 'AL'])
dot.edge('B', 'L', constraint='false')
```

```python
dot
```

```python
dot.edge_attr.update(arrowhead='vee', arrowsize='2')
dot.graph_attr['rankdir'] = 'BT'
dot
```

```python
dot.format = 'svg'
print(dot.pipe().decode('utf-8'))
```

```python
print(dot.source)
```

```python
dot.engine = 'circo'
dot
```

```python
dot.engine = 'neato'
dot
```

```python
import ipywidgets as wdg
from graphviz import Digraph
def add_edge(n1, n2):
    d = Digraph()
    d.edges(['12', '23', '34', '45'])
    d.edge(n1, n2)
    d.graph_attr['rankdir'] = 'LR'
    return d
nodes = [str(i+1) for i in range(5)]
wdg.interact(add_edge, n1=nodes, n2=nodes)
```
